```python
#Варіант 4
import pandas as pd
import numpy as np
import re
```

```python
# Завдання 1: замінити пропущені значення  «-» в таблиці на NaN
url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"
data = pd.read_html(url, header=0)[2]
df1 = data.head(5)
print(df1)
```

```python
# Завдання 2: визначте кількість рядків і стовпчиків за допомогою атрибуту shape
rows, columns = data.shape
print("Рядків: ", rows, " Стовпичків: ", columns)
```

```python
# Завдання 3: визначте загальну кількість років та кількість місяців у даних
years = len(data['Рік']) -1
months = years * 12
print( "Років: ", years, " Місяців: ", months)
```

```python
# Завдання 4: замінити пропущені значення  «-» в таблиці на NaN
data = data.replace("—", np.nan)
data
```

```python
# Завдання 5: визначити тип стовпців за допомогою dtypes
column_types = data.dtypes
print("Типи стовпців: ")
print(column_types)
```

```python
# Завдання 6: видаліть передостанній і останній стовпчики ( Кількість кг хліба[20] на середню зарплату,  Кількість БігМаків на середню зарплату). Далі працюйте з оновленою таблицею
data = data.drop(columns=["Кількість кг хліба[20] на середню зарплату", "Кількість БігМаків на середню зарплату"])
data
```

```python
# Завдання 7: замінити тип нечислових даних на числові
data["Рік"] = data["Рік"].str.replace("*", "")
data["Рік"] = data["Рік"].astype(int)
data["В національній валюті"] = data["В національній валюті"].str.replace('▲', "").str.replace('▼', "")
data["В доларах США"] = data["В доларах США"].str.replace('▲', "").str.replace('▼', "")
data['В національній валюті'] = data['В національній валюті'].str.replace(',', '.').astype(float)
data['В доларах США'] = data['В доларах США'].str.replace(',', '.').astype(float)
print(data)
data.dtypes
```

```python
# Завдання 8: порахуйте частку пропусків в кожній колонці (за допомогою методів isnull , sum, len)
print("Пропуски в колонці років: ", sum(data['Рік'].isnull()))
print("Пропуски в колонці національної валюти: ", sum(data['В національній валюті'].isnull()))
print("Пропуски в колонці доларів США: ", sum(data['В доларах США'].isnull()))
```

```python
# Завдання 9: замініть всі пропущені значення на середні значення по (стовпцю)
data['Рік'].replace(np.nan, data['Рік'].mean())
data['В національній валюті'].replace(np.nan, data['В національній валюті'].mean())
data['В доларах США'].replace(np.nan, data['В доларах США'].mean())
data_graph = data
print(data)
```

```python
# Завдання 10: знайдіть рік з найвищою зарплатою в Україні в грн і вкажіть цю зарплату

print("Рік з найбільшою зарплатою: ", data.loc[data["В національній валюті"].idxmax(), ["Рік"]], " Найбільша зарплата: ", data["В національній валюті"].max())
```

```python
# Завдання 11: знайдіть рік з мінімальною  зарплатою в Україні в дол. США і вкажіть цю зарплату
print("Рік з найменшою зарплатою: ", data.loc[data["В доларах США"].idxmin(), ["Рік"]], " Найменша зарплата: ", data["В доларах США"].min())
```

```python
# Завдання 12: виведіть роки, коли зарплата в грн була вище середньої
UAH_average = np.round(data["В національній валюті"].mean(), 2)
print("Середня зарплата у грн: ", UAH_average)
print("Роки з зарплатою, вищою за середню:")
print(data.loc[data[data["В національній валюті"].gt(UAH_average)].index[0:len(data["Рік"])], ["Рік"]])
```

```python
# Завдання 13: порівняйте зарплату в грн за останні два доступних роки
ind1 = np.argmax(data["Рік"])
ind2 = ind1 - 1
print("Останні доступні роки: ", data["Рік"].iloc[ind2], data["Рік"].iloc[ind1])
print("Зарплата за останні доступні роки: ", data["В національній валюті"].iloc[ind2], data["В національній валюті"].iloc[ind1])
sal22 = data["В національній валюті"].iloc[ind2]
sal23 = data["В національній валюті"].iloc[ind1]
if data["В національній валюті"].iloc[ind2] > data["В національній валюті"].iloc[ind1]:
    print("Зарплата за 2022 рік вища на ", np.round(sal22 - sal23, 2),  " грн.")
elif data["В національній валюті"].iloc[ind2] < data["В національній валюті"].iloc[ind1]:
    print("Зарплата за 2023 рік вища на ", np.round(sal23 - sal22, 2), " грн.")
elif data["В національній валюті"].iloc[ind2] == data["В національній валюті"].iloc[ind1]:
    print("Зарплата за 2022 і 2023 роки однакова.")
```

```python
# Завдання 14: створіть графік, що відображає зарплату в Україні з часом (за роками)
data_graph.plot.bar(x = "Рік", y = "В національній валюті") #Усі колонки окрім 1993, 1994, 1995 мають занадто малі значення, щоб їх було видно поряд з цими трьома на графіку
```

```python
# Завдання 15: видаліть останній рік з таблиці та виведіть оновлену таблицю
df2 = data_graph
df2 = df2.head(-1)
print(df2)
```

```python
# Завдання 16: побудуйте графік, що відображає середню зарплату в Україні в розрізі років без останнього року
df2.plot.bar(x = "Рік", y = "В національній валюті")
```

```python
# Завдання 17: обчисліть зарплату за останні два роки і порівняйте її із зарплатою за останні 5 років. Вкажіть, чи є різниця, і в якому напрямку.
sal_last2 = np.round((df2["В національній валюті"].iloc[ind1-1] + df2["В національній валюті"].iloc[ind1-2])/2, 2)
sal_last5 = np.round((df2["В національній валюті"].iloc[ind1-1] + df2["В національній валюті"].iloc[ind1-2] + df2["В національній валюті"].iloc[ind1-3] + df2["В національній валюті"].iloc[ind1-4] + df2["В національній валюті"].iloc[ind1-5])/5, 2)
print("Середня зарплата за останні 2 роки: ", sal_last2 , " грн.", " Середня зарплата за останні 5 років: ", sal_last5, " грн.")
if sal_last2 > sal_last5:
    print("Зарплата за останні 2 роки вища на ", np.round(sal22 - sal23, 2),  " грн.")
elif sal_last5 < sal_last2:
    print("Зарплата за останні 5 років вища на ", np.round(sal23 - sal22, 2), " грн.")
elif sal_last2 == sal_last5:
    print("Зарплата за останні 2 і останні 5 років однакова.")
```
