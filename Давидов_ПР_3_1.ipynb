```python
#Варіант 4
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Завдання 1: Загрузити датасет   Kaggle : Global YouTube Statistics 2023 за посиланням
url = "https://www.kaggle.com/datasets/nelgiriyewithana/global-youtube-statistics-2023?resource=download"
data = pd.read_csv('D:/Учеба/Машинне Навчання/Lab3/GlobalYouTubeStatistics.csv',  encoding = "ISO-8859-1")
print(data)
```

```python
#Завдання 2: виведіть перші п'ять рядків (використовується метод head):
data.head(5)
```

```python
#Завдання 3: виведіть розміри датасету (використовуйте метод shape):
data.shape
```

```python
#Завдання 4: перевірте, чи у всіх рядків вистачає даних:
#виведіть кількість пропусків  у кожному зі стовпців (використовуйте методи isna та sum).
#Якщо є пропуски, замінить їх на NaN, потім тип даних на числовий.
print("Частка пропусків у ютуберах: ", sum(data['Youtuber'].isnull()))
print("Частка пропусків у підписниках: ", sum(data['subscribers'].isnull()))
print("Частка пропусків у переглядах: ", sum(data['video views'].isnull()))
print("Частка пропусків у категоріях: ", sum(data['category'].isnull()))
print("Частка пропусків у назвах: ", sum(data['Title'].isnull()))
print("Частка пропусків у завантаженнях: ", sum(data['uploads'].isnull()))
print("Частка пропусків у країнах: ", sum(data['Country'].isnull()))
print("Частка пропусків у абревіатурах: ", sum(data['Abbreviation'].isnull()))
print("Частка пропусків у типах каналів: ", sum(data['channel_type'].isnull()))
print("Частка пропусків у підписниках за останні 30 днів: ", sum(data['subscribers_for_last_30_days'].isnull()))
print("Частка пропусків у році створення: ", sum(data['created_year'].isnull()))
print("Частка пропусків у місяці створення: ", sum(data['created_month'].isnull()))
print("Частка пропусків у даті створеення: ", sum(data['created_date'].isnull()))
print("Частка пропусків у освіті: ", sum(data['Gross tertiary education enrollment (%)'].isnull()))
print("Частка пропусків у популяції: ", sum(data['Population'].isnull()))
print("Частка пропусків у безробітності: ", sum(data['Unemployment rate'].isnull()))
print("Частка пропусків у популяції міст: ", sum(data['Urban_population'].isnull()))
print("Частка пропусків у широті: ", sum(data['Latitude'].isnull()))
print("Частка пропусків у довготі: ", sum(data['Longitude'].isnull()))
data = data.replace("—", np.nan)
```

```python
#Завдання 5: замініть комірки з пропущеними значенями на середні за стовпцем. І тип даних на float.
data['subscribers_for_last_30_days'].astype(float)
data['subscribers_for_last_30_days'].fillna(data['subscribers_for_last_30_days'].mean(), inplace=True)
data['created_year'].replace(np.nan, data['created_year'].mean(), inplace=True)
data['created_year'].astype(float)
data['created_date'].replace(np.nan, data['created_date'].mean(), inplace=True)
data['created_date'].astype(float)
data['Gross tertiary education enrollment (%)'].replace(np.nan, data['Gross tertiary education enrollment (%)'].mean(), inplace=True)
data['Gross tertiary education enrollment (%)'].astype(float)
data['Population'].replace(np.nan, data['Population'].mean(), inplace=True)
data['Population'].astype(float)
data['Unemployment rate'].fillna(data['Unemployment rate'].mean(), inplace=True)
data['Unemployment rate'].astype(float)
data['Urban_population'].replace(np.nan, data['Urban_population'].mean(), inplace=True)
data['Urban_population'].astype(float)
data['Latitude'].replace(np.nan, data['Latitude'].mean(), inplace=True)
data['Latitude'].astype(float)
data['Longitude'].replace(np.nan, data['Longitude'].mean(), inplace=True)
data['Longitude'].astype(float)
data.head(10)
```

```python
#Завдання 6: ще раз перевірте кількість пропусків.
print("Частка пропусків у ютуберах: ", sum(data['Youtuber'].isnull()))
print("Частка пропусків у підписниках: ", sum(data['subscribers'].isnull()))
print("Частка пропусків у переглядах: ", sum(data['video views'].isnull()))
print("Частка пропусків у категоріях: ", sum(data['category'].isnull()))
print("Частка пропусків у назвах: ", sum(data['Title'].isnull()))
print("Частка пропусків у завантаженнях: ", sum(data['uploads'].isnull()))
print("Частка пропусків у країнах: ", sum(data['Country'].isnull()))
print("Частка пропусків у абревіатурах: ", sum(data['Abbreviation'].isnull()))
print("Частка пропусків у типах каналів: ", sum(data['channel_type'].isnull()))
print("Частка пропусків у підписниках за останні 30 днів: ", sum(data['subscribers_for_last_30_days'].isnull()))
print("Частка пропусків у році створення: ", sum(data['created_year'].isnull()))
print("Частка пропусків у місяці створення: ", sum(data['created_month'].isnull()))
print("Частка пропусків у даті створеення: ", sum(data['created_date'].isnull()))
print("Частка пропусків у освіті: ", sum(data['Gross tertiary education enrollment (%)'].isnull()))
print("Частка пропусків у популяції: ", sum(data['Population'].isnull()))
print("Частка пропусків у безробітності: ", sum(data['Unemployment rate'].isnull()))
print("Частка пропусків у популяції міст: ", sum(data['Urban_population'].isnull()))
print("Частка пропусків у широті: ", sum(data['Latitude'].isnull()))
print("Частка пропусків у довготі: ", sum(data['Longitude'].isnull()))
```

```python
#Завдання 9: визначте максимальну, мінімальну, середню кількість переглядів (використовуйте методи max, min, mean)
print("Максимальна кількість переглядів", round(data['video views'].max(), 0))
print("Мінімальна кількість переглядів", round(data['video views'].min(), 0))
print("Середня кількість переглядів", round(data['video views'].mean(), 0))
```

```python
#Завдання 10: знайдіть країну, де найбільше відео було завантажено на YouTube
print("Країна, з якої завантажено найбільше відео: ", data.loc[data["uploads"].idxmax(), ["Country"]])
```
